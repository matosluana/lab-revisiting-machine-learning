```python
df = pd.DataFrame(categorical['STATE'].value_counts()).reset_index()

df.columns = ['state', 'count']
other_states = list(df[df['count']<2500]['state'])

def clean_state(x):
    if x in other_states:
        return 'other'
    else:
        return x
    
categorical['STATE'] = list(map(clean_state, categorical['STATE']))
```

```python
### I will predict some columns, but before doing that I will fill the NaN where possible (CLUSTER2)
numerical['CLUSTER2'].value_counts()
```

```python
# Filling with median
numerical['CLUSTER2'] = numerical['CLUSTER2'].fillna(np.ceil(np.mean(numerical['CLUSTER2'])))
```

```python
# I will predict them using models
# Since I have more than one colum to predict I will create a function

def prediction_model(feature):
    numerical = pd.read_csv('numerical.csv')
    numericals_with_feature = numerical[numerical[feature].notna()]
    #dropping feature and all columns which still contain NaN
    X_feature = numericals_with_feature.drop(['INCOME','WEALTH1', 'WEALTH2','MSA','ADI','DMA','NEXTDATE','TIMELAG'],axis=1)
    y_feature = numericals_with_feature[feature]
    
    X_normalized_feature = pd.DataFrame(MinMaxScaler().fit_transform(X_feature), columns=X_feature.columns)
        
    KNN = KNeighborsClassifier(n_neighbors=3)
    KNN.fit(X_normalized_feature,y_feature)
    
    X_all = numerical.drop(['INCOME','WEALTH1', 'WEALTH2','MSA','ADI','DMA','NEXTDATE','TIMELAG'],axis=1)
    X_normalized_all = pd.DataFrame(MinMaxScaler().fit_transform(X_all), columns=X_all.columns)
    y_all = numerical[feature]
    
    name = 'predicted_'+feature
    predicted_feature = pd.Series(KNN.predict(X_normalized_all),name=name)
    
    return predicted_feature
```

```python
numerical.to_csv(r'C:\Users\luana\IH-Labs\lab-revisiting-machine-learning\numerical.csv', index=False)
```

```python
def replacement_nan(feature):
    numerical = pd.read_csv('numerical.csv')
    predicted_feature = prediction_model(feature)
    numerical = pd.concat([numerical, predicted_feature], axis=1)
    numerical[feature] =  np.where(numerical[feature].isna(), predicted_feature, numerical[feature])
    
    return numerical[feature]
```

```python
%%time
to_predict = ['INCOME','WEALTH1', 'WEALTH2','MSA','ADI','DMA']

for item in to_predict:
    numerical[item]= replacement_nan(item)
```
